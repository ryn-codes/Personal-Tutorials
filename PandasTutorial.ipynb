```python
df = pd.read_csv('deliveries.csv')
df.head()
```

```python
# Top 10 batsman by batsman_runs
df.groupby('batsman')[['batsman','batsman_runs']].sum().sort_values(ascending=False, by='batsman_runs').head(10)
```

```python
df.columns
```

```python
# Top 10 batsman who hit maximum 6s
# df[df['batsman_runs'] == 6] # new dataframe
df[df['batsman_runs'] == 6].groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(10)
```

```python
# Virat Kohli scoring maximum runs against which 3 teams?
df[df['batsman'] == 'V Kohli'].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
def runs_scored(batsman_name):
  return df[df['batsman'] == batsman_name].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
runs_scored('DA Warner')
```

```python
runs_scored('G Gambhir')
```

## isin() function

```python
# Find the most destructive death over batsman in the history of IPL
# Strike rate = (no of runs/no of balls)*100
# Min batsman 200 balls in over 16-20
```

```python
death_overs = df[df['over'] > 15]
death_overs.groupby('batsman')['batsman_runs'].count()
```

```python
all_batsman = death_overs.groupby('batsman')['batsman_runs'].count()
x = all_batsman > 200
x
```

```python
all_batsman[x]
```

```python
batsman_list = all_batsman[x].index.tolist()
```

```python
# runs scored by all these 43 batsman in death overs
# balls played by these 43 batsman in death overs

death_overs['batsman'].isin(batsman_list)
```

```python
final = death_overs[death_overs['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
sr = (runs/balls)*100
sr.sort_values(ascending=False).head(10)
```
